```python
## Matplotlib is a graphing and plotting library for Python
## A matrix is a two dimensional data structure or a list of lists in which numbers are arranged into rows and columns
## Pyplot provides a MATLAB-like interface for making plots

# import dependency; chain matplotlib with pyplot and then refer to it as a shortened alias
import matplotlib.pyplot as plt

# checks current version
# matplotlib.__version__
```

```python
# % sign is called a magic command
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot using matlab method
# chain plot() function to plt and add the above variables
plt.plot(x_axis, y_axis)
```

```python
## fig and ax are variables 
## plt.subplot() is the function that returns a tuple (which stores multiple items in a single variable)
fig, ax = plt.subplots()

# create plot with ax.plt()
## ax.plot is used when we switch from MATLAB to object-oriented interface method
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
## object oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
## plt.show() functions looks for all acctive figure objects and opens a window that displays them if you have two or more sets of data to plot
## use function once per session as it can cause graphs to display improperly
## restart 
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# create plot and add a lbel for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45
plt.ylim (0, 45)
plt.xlim ("Jan", "June")

# create a title
plt.title("PyBer Fare by Month")

# add a grid
plt.grid()

# add legend
plt.legend()
```

```python
# Matplotlib Functions	Feature
# plt.figure(figsize=(w, h))	Change the size of the figure in pixels. Added on the first line of the script.
# plt.plot(x, y, label='line')	Add a label that will be added to the legend.
# plt.xlim(min, max)	Set the min and max range of the x-axis.
# plt.ylim(min, max)	Set the min and max range of the y-axis.
# plt.xlabel('x label')	Add a label to the x-axis.
# plt.ylabel('y label')	Add a label to the y-axis.
# plt.title("Title")	Add a title.
# plt.legend()	Add a legend.
# plt.grid()	Add a grid to the chart.
# plt.savefig("add a path and figure extension")	Save the figure with the given extension. Added at the end of the script.
```

```python
# create plot and add a lbel for the legend
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")

# create labels for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# y limit
ax.set_ylim(0, 45)

# create a title
ax.set_title("PyBer Fare by Month")

# add a grid
ax.grid()

# add legend
ax.legend()
```

```python
# create bar graph
plt.bar(x_axis, y_axis, color="red", label="Boston")

# create labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

# create a title
plt.title("PyBer Fare by Month")

# add the legend
plt.legend()
```

```python
# create a horizontal bar graph
## gca() means get current axes; .invert is chaained to switch x and y axes
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fares($)")
plt.title('Fares per Month')
```

```python
# vertical bar graph
## fig, ax variables need to be set first

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label="New York")
```

```python
# object oriented h bar graph

fig, ax = plt.subplots()

ax.barh(x_axis, y_axis, color="Cyan", label="Chi Town")
ax.legend()
ax.set_xlabel("Fares($)")
ax.set_ylabel("Month")
ax.invert_yaxis()
```

```python
## list comprehension technique 
## basically multiyping each data set (i) by a number (3)
## plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])


# defining a new variable to adjust size of markers times 3
y_axis_large = []
for data in y_axis:
    y_axis_large.append(data*3)

# adding "o" inside plt.plot() switches the function from a line chart to a scatter plot
# alt you can use plt.scatter()
# s changes the size of the marker as it relates to the number; bigger the number, bigger the dot

plt.scatter(x_axis, y_axis, s=y_axis_large, color="red", label="Chicago")
plt.ylabel("Fare($)")
plt.xlabel("Month")
plt.ylim(0, 50)
plt.title("PyBer Fare by Month")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color = "skyblue", linewidth = 2, label = "Boston")
ax.legend()
ax.set_title = "PyBer Bubbles"
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.set_xlim(0, 50)

## need to figure out how add borders to markers and transparency
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow=True,
    startangle=270
       )
plt.gca().invert_yaxis()
plt.show()
```
